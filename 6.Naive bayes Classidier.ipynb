### Step-1: Call libraries

```python
import pandas as pd
from google.colab import files
import io
from sklearn.naive_bayes import GaussianNB
```

#### step-2: Read the data

```python
uploaded = files.upload()

data = pd.read_csv(io.BytesIO(uploaded['Iris_data.csv']))
data
```

```python
x = data.values[:,0:4]
x
```

```python
y = data.values[:,4]
y
```

### Step=3: Develop model

```python
model = GaussianNB()

model.fit(x,y)
```

```python
pred = model.predict(x)
pred
```

```python
table = pd.crosstab(y,pred)
table
```

Accuracy


```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y,pred)*100  #OR model.score(test_y,pred_test) for accuracy
accuracy
```

Misclassification_error

```python
Misclassification_error = 100-accuracy
Misclassification_error
```
