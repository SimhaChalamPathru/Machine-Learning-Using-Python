# Descriptive Statistics

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
import pandas as pd
data=pd.read_csv(io.BytesIO(uploaded['Credit_Card_Expenses.csv']))
data
```

```python
data.head(5) # shows the first 5 examples of data
```

```python
# Read CC_Expanses to CC

cc=data.CC_Expenses
cc
```

```python
# mean

cc.mean()
```

```python
# Median

cc.median()
```

```python
# Mode

cc.mode()
```

```python
# Standard Deviation 

cc.std()
```

```python
# Variance

cc.var()
```

```python
# Minimum

cc.min()
min(cc)
```

```python
# Maximum

cc.max()
max(cc)
```

```python
# Percentile

cc.quantile()
```

```python
# Skewness

cc.skew()
```

```python
# Kurtosis

cc.kurt()
```

```python
# count

cc.count()
len(cc)
```

```python
# sum

cc.sum()
sum(cc)
```

```python
# product

cc.prod()
```

```python
# Square root

import math as mymath
mymath.sqrt(49)
```

```python
# Sum of squares

sum(cc**2)
```

```python
# Descriptive Statistics

cc.describe()
```

```python
# Graphs

# Histogram

import matplotlib.pyplot as myplot
myplot.hist(cc)
myplot.show()

# Box plot

myplot.boxplot(cc)
myplot.show()
```
