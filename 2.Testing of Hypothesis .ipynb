# One- Sample t-test:


```python
"""A company claims that on an average it takes only 40 hours to
process any purchase order. Based on the data given below, can you validate
the claim? The data is given in PO_Processing.csv
"""

#import pandas as pd
#data = pd.read_csv("https://raw.githubuser")

       # OR
from google.colab import files
uploaded = files.upload()
import io
import pandas as pd
data = pd.read_csv(io.BytesIO(uploaded['PO_Processing.csv']))
data
```

```python
PT = data.Processing_Time
PT
```

```python
# Performing one-sample t-test
import scipy
from scipy import stats 
stats.ttest_1samp(PT,40)
```

# Two sample t-test


```python
"""A super market chain has introduced a promotional activity in its
selected outlets in the city to increase the sales volume. Based on the data
given below, check whether the promotional activity resulted in increasing the
sales. The outlets where promotional activity introduced are denoted by 1 and
others by 2? The data is given in Sales_Promotion.csv
"""

```

```python
from google.colab import files
uploaded = files.upload()
import io
import pandas as pd
from scipy import stats
data=pd.read_csv(io.BytesIO(uploaded['Sales_Promotion.csv']))
data
```

```python
# Reading Variables

s1 = data.Sales_Out1
s2 = data.Sales_Out2
s1,s2
```

```python
# 2 sample t test

stats.ttest_ind(s1,s2)
```

# Normality Tests

```python
"""The processing times of purchase orders is given in
PO_Processing.csv. Is processing time normally distributed?
"""
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
import pandas as pd
from scipy import stats
import matplotlib.pyplot as myplot
data = pd.read_csv(io.BytesIO(uploaded['PO_Processing.csv']))
data
```

```python
PT = data.Processing_Time
PT
```

### Normality Checking using Q-Q plot

```python
stats.probplot(PT,plot=myplot)
myplot.show()
```

### Normality Checking using Shapiro - wilk test


```python
stats.mstats.normaltest(PT)
```

# Analysis of variance

```python
# One Way ANOVA


"""An electronics and home appliance chain suspect the location of shelves
where television sets are kept will influence the sales revenue. The data on
sales revenue in lakhs from the television sets when they are kept at
different locations inside the store are given in sales revenue data file. The
location is denoted as 1:front, 2: middle & 3: rear. Verify the doubt? The data
is given in Sales_Revenue_Anova.csv."""
```

```python
# Importing Packages 

from google.colab import files
import io
import pandas as pd
from scipy import stats
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
# import data

uploaded = files.upload()
data=pd.read_csv(io.BytesIO(uploaded['Sales_Revenue_Anova.csv']))
data
sales=data.Sales_Revenue
location = data.Location
```

```python
# Computing ANOVA table

mymodel = ols('sales~C(location)',data).fit()
mymodel
anova_table = anova_lm(mymodel)
anova_table
```

```python
#location wise means

sales.groupby(location).mean()
```

```python
# Box plot location Wise

import matplotlib.pyplot as plt
data.boxplot(column='Sales_Revenue',by='Location')
myplot.show()
```
