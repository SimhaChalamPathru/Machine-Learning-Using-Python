# K-nearest Neighbours

Develop a model to predict the iris plant class (1: Iris-setosa, 2: Irisversicolor & 3: Iris-virginica) based on sepal length, sepal width, petal length and
petal width using Na√Øve Bayes classifier. The data is given in Iris_data.csv file.
Validate the model with Iris_test.csv data? Use k = 5

Step-1: Call libraries 


```python
import pandas as pd
import io
from google.colab import files
from sklearn.neighbors import KNeighborsClassifier
```

Step-2: Import Data

```python
uploaded = files.upload()
data = pd.read_csv(io.BytesIO(uploaded['Iris_data.csv']))
data
```

```python
x = data.values[:,0:4]
x
```

```python
y = data.values[:,4]
y
```

Develop the model

```python
model = KNeighborsClassifier(n_neighbors = 5)
model.fit(x,y)
```

Accuracy

```python
Accuracy = model.score(x,y)  # it for Accuracy
Accuracy
```

misclassification_error

```python
misclassification_error = 1 - Accuracy
misclassification_error
```

## validating the model on Test data

```python
uploaded = files.upload()
test_data = pd.read_csv(io.BytesIO(uploaded['Iris_test.csv']))
test_data
```

```python
test_x = test_data.values[:,0:4]
test_x
```

```python
test_y = test_data.values[:,4]
test_y
```

```python
test_pred = model.predict(test_x)
test_pred
```

```python
test_table = pd.crosstab(test_y,test_pred)
test_table
```

Accuracy

```python
accuracy = model.score(test_x,test_y)
accuracy
```

miscassification_error

```python
miscassification_error = 1- accuracy
miscassification_error
```
