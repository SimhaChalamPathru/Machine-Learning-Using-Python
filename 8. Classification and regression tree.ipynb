# Classification and regression tree

A marketing company wants to optimize their mailing campaign by sending
the brochure mail only to those customers who responded to previous mail
campaigns. The profile of customers are given below. Can you develop a rule to
identify the profile of customers who are likely to respond (Mail_Respond.csv)?

step-1: Import Packages

```python
import pandas as pd
from sklearn import tree
from google.colab import files
import io
```

Step-2: read files and variables

```python
uploaded = files.upload()

data = pd.read_csv(io.BytesIO(uploaded["Mail_Respond.csv"]))
data
```

```python
x = data[["District","House_Type","Income","Previous_Customer"]]
x
```

```python
y = data.Outcome
y
```

Step=3: Develop the model

```python
model = tree.DecisionTreeClassifier(min_samples_split = 10)
model.fit(x,y)
print(model.score(x,y))  # accuracy
1-model.score(x,y)  # Misclassification error
```

Model Accuracy measures

```python
pred = model.predict(x)
pred
```

```python
table = pd.crosstab(y,pred)
table
```
